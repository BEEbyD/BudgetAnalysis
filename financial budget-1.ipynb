top priorities of the government for the financial year:

```python
data.plot(x='Department /Ministry', y='Fund allotted(in ₹crores)',kind='bar',figsize=(13,6),rot=45 )
plt.ylabel("Fund allotted in '000,000")
plt.title("Budget allocation in India",fontsize=20)

 #data.plot(kind='bar')
```

Finance department is getting the most of the share from the total budget of the government of india.  

```python
##donut plot to have a clear view of the distribution of funds among all the departments:
df = data["Fund allotted(in ₹crores)"]
labels = data["Department /Ministry"]
plt.figure(figsize=(7,11))
plt.pie(df, labels=labels, autopct='%1.1f%%', startangle=90, pctdistance=0.85, shadow =True)
central_circle = plt.Circle((0, 0), 0.5, color='white')
fig = plt.gcf()
fig.gca().add_artist(central_circle)
plt.rc('font', size=11)
plt.title("Distribution of The Budget", fontsize=20)
plt.show()
```

#Conclusion#
Finance department is getting 40% of the funds.
